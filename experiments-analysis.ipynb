```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
exp_l256 = pd.read_csv("DATA/experiments/Lublin_256_runtimes.csv")
exp_l1024 = pd.read_csv("DATA/experiments/Lublin_1024_runtimes.csv")
```

```python
def lublin_comparison_plot(data_l256, data_l1024, xlim):
    fig, axs = plt.subplots(figsize=(16, 8), nrows=2, ncols=1)

    fig.suptitle("Lublin workload with 256 and 1024 processors")

    for ax in axs.flat:
        ax.set(xlabel='Average Bounded Slowdown', ylabel='Scheduling policies')
        ax.set_xlim(xlim)
    
    sns.boxplot(
        ax=axs[0],
        data=data_l256,
        orient="h",
        palette="Set3",
        marker=".")
    
    sns.boxplot(
        ax=axs[1],
        data=data_l1024,
        orient="h",
        palette="Set3",
        marker=".")
    
    plt.show()
```

```python
xlim = (-1000,22000)
lublin_comparison_plot(exp_l256, exp_l1024, xlim)
```
